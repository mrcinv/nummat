# Minimalne ploskve
Žično zanko potopimo v milnico. Opna, ki se ustvari je oblike, ki ima med vsemi možnimi ploskvami minimalno površino. Takim ploskvam rečemo *minimalne ploskve*. Če ploskev predstavimo kot graf funkcije 2 spremenljivk

$$z = F(x,y)$$

potem minimalna ploskev zadošča Laplaceovi enačbi

$$\triangle F(x,y)=\frac{\partial^2}{\partial x^2}F+\frac{\partial^2}{\partial y^2}F=0.$$

## Minimalne ploskve na pravokotni zanki

V tej vaji bomo iskali minimalno ploskev, ki je dana z grafom funkcije na pravokotniku $[a,b]\times[c,d]$ in ima predpisane vrednosti na robu. Problem bomo diskretizirali, tako da bomo iskali vrednosti le v končno mnogo točkah $(x_i,y_j)$, kjer je 

$$x_i=a+ih\quad y_j=c+ik$$

za $h=\frac{b-a}{n}$ in $k=\frac{d-c}{m}$ in $n,m\in\mathbb{N}$. 

Če sta $h$ in $k$ dovolj majhna, lahko brez škode predpostavimo, da lahko izberemo $n$ in $m$, da je $h=k$.

```python
%pylab inline
fig, ax = subplots()
x = linspace(0,1,10)
X, Y = meshgrid(x,x)
ax.plot(X.reshape(100,1),Y.reshape(100,1),'x')
ax.annotate("$(x_i,y_j)$", xy=(x[3], x[4]), xytext=(5,5),
            textcoords='offset points', ha='left', va='bottom')
title("To\\v cke v katerih ra\\v cunamo vrednosti F(x,y)")
show()
```

Iščemo torej vrednosti 

$F_{i,j}=F(x_i,y_j).$

Ker so vrednosti predpisane na robu, so tudi vrednosti $F_{0j}$, $F_{i0}$, $F_{nj}$ in $F_{im}$ predpisane. 
## Diskretizacija Laplaceove enačbe
Če uporabimo končne diference, lahko Laplaceov operator približno izrazimo z vrednostmi $F_{ij}$

$$ \triangle F(x_i,y_j)=\frac{F_{i-1j}-2F_{ij}+F_{i+1j}}{h^2}+\frac{F_{ij-1}-2F_{ij}+F_{ij+1}}{h^2}=\frac{F_{i-1j}+F_{ij-1}-4F_{ij}+F_{i+1j}+F_{ij+1}}{h^2}$$

Tako dobimo sistem enačb

$$ \frac{F_{i-1j}+F_{ij-1}-4F_{ij}+F_{i+1j}+F_{ij+1}}{h^2}=0,$$

ki ga bomo rešili z *Jacobijevo*, *Gauss-Seidlovo* in *SOR* iteracijo. Ideja iteracijskih metod je, da rešitev poiščemo z rekurzivno formulo. V našem primeru lahko iz enačbe izrazimo diagonalni element $F_{ij}$ in dobimo preprosto rekurzivno enačbo:

$$ F^{(n+1)}_{ij} = \frac{F^{(n)}_{i-1j}+F^{(n)}_{ij-1}+F^{(n)}_{i+1j}+F^{(n)}_{ij+1}}{4}$$

ki ni nič drugega kot *Jacobijeva iteracija*.

```python
F=zeros((20,20))
# na robu predpišemo vrednosti
x=linspace(0,pi,20)
F[:,0]=sin(x)
X, Y = meshgrid(x,x)
```

```python
%pylab inline
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_wireframe(X,Y,F)
```

```python
%load minimal.py
```

```python
# Jacobijeva iteracija
n,m = F.shape
Fp = F.copy()
napaka = 1
k = 0
maxk = 1000
tol = 0.0001
while (napaka > tol) and (k<maxk):
    # en korak Jacobijeve iteracije
    for i in range(1,n-2):
        for j in range(1,m-2):
            F[i,j]=(Fp[i-1,j]+Fp[i+1,j]+Fp[i,j-1]+Fp[i,j+1])/4
    napaka = norm(F-Fp)
    Fp = F.copy()
    k = k+1
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_wireframe(X,Y,F)
```

## Gauss-Seidlova iteracija
Gauss-Seidlovo iteracijo dobimo, tako da v Jacobijevi iteraciji uporabimo nove približke, ki smo jih do tistega trenutka že izračunali.

$$ F^{(n+1)}_{ij} = \frac{F^{(n+1)}_{i-1j}+F^{(n+1)}_{ij-1}+F^{(n)}_{i+1j}+F^{(n)}_{ij+1}}{4}$$

Program je toliko bolj preprost, saj lahko delamo z eno samo matriko.

```python
F=zeros((20,20))
# na robu predpišemo vrednosti
x=linspace(0,pi,20)
F[0,:]=sin(x)
X, Y = meshgrid(x,x)

n,m = F.shape
Fp = F.copy()
napaka = 1
k = 0
maxk = 1000
tol = 0.0001
while (napaka > tol) and (k<maxk):
    # en korak Gauss Seidel iteracije
    for i in range(1,n-2):
        for j in range(1,m-2):
            # GS sprogramiramo tako, da uporabimo v formuli isti 
            # vektor(matriko v našem primeru)
            F[i,j]=(F[i-1,j]+F[i+1,j]+F[i,j-1]+F[i,j+1])/4
    napaka = norm(F-Fp)
    Fp = F.copy()
    k = k+1
```

## SOR iteracija

Pri SOR (successive Over Relaxation) iteraciji poskušamo izboljšati hitrost konvergence, tako da približek po Gauss-Seidlu malce zmešamo z enim približkom prej

$$ F^{(n+1)}_{ij} = \omega\left(\frac{F^{(n+1)}_{i-1j}+F^{(n+1)}_{ij-1}+F^{(n)}_{i+1j}+F^{(n)}_{ij+1}}{4}\right)+(1-\omega)F^{(n)}_{ij}$$

Hitrost konvergence je odvisna od parametra $\omega$

```python
X, Y = meshgrid(x,x)
omega_list = arange(1,2,0.05)
koraki = []
for omega in omega_list:
    F=zeros((20,20))
    # na robu predpišemo vrednosti
    x=linspace(0,pi,20)
    F[0,:]=sin(x)
    n,m = F.shape
    Fp = F.copy()
    napaka = 1
    k = 0
    maxk = 1000
    tol = 0.01
    while (napaka > tol) and (k<maxk):
        # en korak SOR iteracije
        for i in range(1,n-2):
            for j in range(1,m-2):
                # približek pri GS „zmešamo“ z prejšnjim približkom
                F[i,j]=omega*(F[i-1,j]+F[i+1,j]+F[i,j-1]+F[i,j+1])/4-(1-omega)*F[i,j]
        napaka = norm(F-Fp)
        Fp = F.copy()
        k = k+1
    koraki.append(k)
```

```python
plot(omega_list,koraki,'*')
title("\v Stevilo korakov SOR v odvisnosti od $\omega$")
show()
```
