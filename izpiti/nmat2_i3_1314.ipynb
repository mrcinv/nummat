Eliptični integral je dan s formulo
  \begin{equation}
    \label{eq:int}
    K(\phi,m) = \int_0^\phi\frac{dt}{\sqrt{1-m\sin^2t}}.
  \end{equation}
    1. S sestavljenim Simpsonovim pravilom izračunaj približek za $K(0.5,2)$ na 5 decimalk natančno.  
    2. Eliptična funkcija $y=\sn(x,m)$ je definirana z relacijo
\begin{equation}
\sin(\phi)=\sn\left(K(\phi,m)\right).\label{eq:sn}
\end{equation}
Izračunaj približek za $\sn(0.5,2)$, ki bo imel vsaj 5 pravilnih števk. *Namig: Z eno od 
numeričnih metod za nelinearne enačbe poišči $\phi$, ki reši enačbo *
\begin{equation}
  \label{eq:enacba}
   x = 0.5 = \int_0^\phi\frac{dt}{\sqrt{1-2\sin^2t}}.
\end{equation}


```python
# sestavljeno simpsonovo pravilo
def simpson(a,b,fun,n):
    x = linspace(a,b,2*n+1)
    y = fun(x)
    w = ones_like(x)
    w[1:-1:2]=4
    w[2:-2:2]=2
    return (b-a)/n/6*sum(y*w)
```
