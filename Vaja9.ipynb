# Reševanje navadnih diferencialnih enačb 
Rešujemo diferencialno enačbo oblike

$$y'(x) = f(y(x),x).$$

Množica rešitev je enoparametrična družina funkcij $y(x)$. Če predpišemo še vrednost funkcije v neki točki t. i. *začetni pogoj*

$y(x_0)=y_0$

je rešitev ena sama. Problem iskanja rešitve NDE z začetnim pogojem imenujemo *začetni problem* ali *Cauchyjeva naloga*. Ogledali si bomo preprosto numerično metodo za reševanje začetnega problema imenovano *Eulerjeva metoda*.

Iščemo rešitev NDE $y'=f(y,x)$ z začetnim pogojem $y(x_0)=y_0$ na intervalu $[x_0,x_k]$. Izberimo si končno mnogo točk 

$$x_0< x_1 < \ldots < x_{n+1}=x_k$$

Z Eulerjevo metodo izračunamo zaporedje približkov $y_i$  za $y(x_i)$ z rekurzivno formulo

$$ h_n =  (x_{n+1}-x_n)$$
$$ y_{n+1} = y_n + h_n f(y_n,x_n). $$

Pogosto vzamemo ekvidistančne točke in fiksen korak $h_n=h$.

**Opomba**: Eulerjeva metoda je 1. reda, kar pomeni, da napaka zelo počasi pada (sorazmerno s $h$), ko manjšamo korak $h$. Zato se v praksi redko uporablja, saj obstajajo metode višjih redov (4,5,6), kjer je  

```python
# Eulerjeva metoda za reševanje NDE
def euler(df, t0, tk, y0, n):
    """Funkcija izračuna rešitev začetnega problema za NDE
    y'(t) = df(y,t) z začetnimi pogoji y(t0)=y0 na intervalu [t0,tk]"""
    t = linspace(t0,tk,n+1)
    h = t[1]-t[0]
    y = zeros_like(t)
    y[0] = y0
    for i in range(n):
        y[i+1] = y[i] + h*df(y[i],t[i])
    return (t,y)
```

Poskusimo rešiti začetni problem za DE

$$y'=y\sin(x)$$ 

z začetnim pogojem $y(0)=1$.

```python
# funkcija desnih strani (polje smeri)
f = lambda y,x: y*sin(x)
(t,y) = euler(f, 0, 10, 1, 100)
plot(t,y)
```

```python
plot(t,y-exp(-cos(t)+1))
```

# Red metode
Če rešujemo začetni problem za NDE na fiksnem intervalu $[t_0,t_k]$,  napaka približka pada s potenco $h$

$$ y^(n)-y(t_k) = C h^r $$

Eksponent $r$ v potenci $h^r$ imenujemo *red metode*. Red metode lahko določimo numerično. 

```python
# odvisnost napake od h
n = arange(100,10000,1000)
h = 10./n
napaka = []
for i in range(len(n)):
    (t,y) = euler(f,0.,10,1,n[i])
    napaka.append( max(abs(y-exp(-cos(t)+1))))
```

```python
plot(log10(h),log10(napaka),'*')
title("Napaka v odvisnosti od h v log-log skali")
```

Red metode lahko določimo tako, da izračunamo naklonski koeficient premice, ki aproksimira odvisnost logaritma napake od logaritma $h$. V primeru Eulerjeve metode je red enak 1.

```python
# napaka = C h^r, kjer je r red metode
# red dobimo, tako da izračunamo naklon premice v log-log grafu.
polyfit(log(h),log(napaka),1)
```

#Dinamika telesne teže

Sledili bomo [članku](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC2764961/pdf/nihms-132219.pdf) in izpeljali model, ki bo opisal časovni potek spremembe telesne teže v odvisnosti od vnosa(hrana) in porabe energije(telesna aktivnost, prebava, toplota). Osnova za model je 1. zakon termodinamike za odprt sistem, ki pravi, da je sprememba shranjene energije $R$ enaka razliki med prejetim energijskim tokom $I$ in oddanim energijskim tokom $E$:

$$R=I-E$$

Vsi zelo dobro vemo, da se shranjena energija v telesu nalaga v obliki glikogenskih in maščobnih zalog, kar vpliva na spremebo telesne teže. Spremeba mase telesnih maščob $F$ je sorazmerna s spremembo shranjene energije $R$

$$R = \lambda_i \frac{dF}{dt},$$

kjer je $\lambda_M = 7325\mathrm{kcal/kg}$ (za moške) in $\lambda_Z=7256\mathrm{kcal/kg}$ (za ženske).

Predpostavimo, da je telesna masa $M$ afino odvisna od mase maščobnih zalog $F$

$$M = \alpha F + b,$$

kjer sta $\alpha_M=1.56$ in $\alpha_Z=1.32$ empirično izmerjeni konstanti, ki se razlikujeta med spoloma. 
Sprememba shranjene energije je potem enaka

$$R = \alpha \lambda \frac{dM}{dt}$$

```python
alambda = 1.56*7325. #moški
# alambda = 1.32*7256. #ženske
```

## Poraba energije $E$

Energijsko porabo lahko v grobem razdelimo na 4 dele

$$E = DIT + PA + RMR + NEAT$$

<table>
<tr><td> DIT </td> <td>sproščena toplota ob prebavi </td><td> $DIT=\beta I$</td><td>$0.04<\beta<1.14 $ </td>
<tr><td> PA </td> <td>fizična aktivnost </td><td> $PA=m\cdot M$</td><td>$m$ določena iz podatov</td>
<tr><td> RMR </td> <td>poraba energije v mirovanju</td><td> $RMR=a M^p - b \cdot S$</td><td> $S$ starost v letih, $a_M=293$, $a_Z=248$, $p=0.43$, $b_M=5.1$ in $b_Z=5.9$</td>
<tr><td> NEAT </td> <td>nezavedna fizična aktivnost </td><td> $NEAT=\frac{r}{1-r}\left(DIT + PA + RMR\right) + C$</td><td>$r=2/3$, $C$ konstanta</td>
</table>

Konstanto $C$ določimo iz predpostavke, da je delež $NEAT$ v $E$ približno 10%. Podobno določimo konstanto $m$.

```python
# moški 40 let, 110kg, vnos 3345kcal/dan
a, p, b, S0, M0, I0 = (293., 0.43, 5.1, 40., 110., 3345) 
# ženska 25 let 70kg, vnos 2300
#a, p, b, S0, M0, I0 = (248., 0.43, 5.9, 25., 70., 2300) 
RMR = a*M0**p - b*S0
E = RMR/0.6
DIT = 0.08*E
NEAT = 0.1*E
PA = E-RMR-DIT # -NEAT
print "Poraba v mirovanju %1.2f kcal/dan" % RMR
print "Celotna poraba energije %1.2f kcal/dan" % E
print "Fizična aktivnost %1.2f kcal/dan" % PA
m, beta, C = (PA/M0, DIT/I0, NEAT-2*(PA+DIT+RMR))
print "m = %1.2f, beta = %1.2f, C = %1.2f" % (m, beta, C)
```

Uporabili bomo model brez člena NEAT, saj je konstanto C težko primerno določiti.

```python
# model brez NEAT člena
rmr = lambda t,M: a*M**p - b*(S0+t/365)
dM = lambda M,t: ((1-beta)*I(t) - (m*M + a*M**p - b*(S0+t/365.)))/alambda
```

```python
def simulate(M0,let=5):
    l = 365
    n=let*l
    t1,M1 = euler(dM,0,n,M0,n)
    plot(t1/l,M1)
    t = linspace(0,n,10)
    M = linspace(min(M1),max(M1),10)
    (tt,MM)=meshgrid(t,M)
    fi = arctan(l*poraba(MM,tt))
    quiver(t/l,M,cos(fi),sin(fi),angles='xy', scale_units='x', scale=3)
    title("Sprmemba mase v %d letih"%let)
    figure()
    plot(t1/l,poraba(M1,t1))
    title("Dnevna sprememba mase")
    grid()
    return(t1,M1)
```

## Simulacije

Izvedli bomo simulacije za različne scenarije. Najprej si poglejmo, kako se bo spreminjala teža, če ne bo nobenih sprememb življenskega sloga.
