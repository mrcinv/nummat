## 3. Domača naloga (največ 9)

```python
# Čebiševe točke
x_cheb = lambda n: cos(linspace(0,pi,n))
plot(x_cheb(5),zeros((5,)),'o')
x_cheb(5)
```

```python
# baricentrična formula
def baricentric(x,f,lam,xi):
    """Izračuna vrednosti interpolacijskega polinoma v x z vrednostmi funkcije f v točkah xi"""
    idx = find(x==xi)
    if len(idx)>0:
        return f[idx]
    else:
        return sum(f*lam/(x-xi))/sum(lam/(x-xi))
utezi = lambda n: (-1)**arange(n)*hstack((0.5,ones((n-2,)),0.5))
```

```python
xi=ab_cheb(10,pi,4*pi)
lam = utezi(10)
f=sin(xi)
baricentric(4*pi-1,f,lam,xi)
```

```python
n = 20
fun = lambda x: exp(-x**2)
xi = x_cheb(n)
f = fun(xi)
lam = utezi(n)
plot(xi,fun(xi),'*')
t = linspace(-1,1,200)
p = vectorize(lambda x: baricentric(x,f,lam,xi))
plot(t,fun(t),'b')
title(u"interpolacija na %d točkah"%n)
figure()
plot(t, p(t)-fun(t),'g')
title("napaka interpolacije")
```

```python
ab_cheb = lambda n,a,b: a+(b-a)*(1+cos(linspace(0,pi,n)))/2
# Lagrangeva interpolacija
def int_poly(fun,a,b,n):
    xi = a+(b-a)*(1+cos(linspace(0,pi,n)))/2
    lam = (-1)**arange(n)*hstack((0.5,ones((n-2,)),0.5))
    f = fun(xi)
    return vectorize(lambda x: baricentric(x,f,lam,xi))
```

```python
n,a,b = (20,0,10)
def risi(n,a,b,fun):
    psin = int_poly(sin,a,b,n)
    xi = ab_cheb(n,a,b)
    plot(xi, fun(xi),'*')
    t = linspace(a,b)
    plot(t,psin(t))
    title("Interpolacijski polinom stopnje %d" % (n-1))
    figure()
    plot(t,psin(t)-fun(t))
    title("Napaka interpolacije")
risi(n,a,b,sin)
```
